Latine America user data extraction

```python
#latine America country list (30 Countries) 
#list from Britannica = https://www.britannica.com/topic/list-of-countries-in-Latin-America-2061416)

countries_Latine = [
'Belize','Costa Rica','El Salvador','Guatemala','Honduras','Mexico','Nicaragua',
'Panama','Argentina','Bolivia',
'Brazil','Chile','Colombia','Ecuador',
'French Guiana','Guyana','Paraguay','Peru','Suriname','Uruguay','Venezuela',
'Cuba','Dominican Republic','Haiti','Guadeloupe','Martinique',
'Puerto Rico','Saint-Barthélemy','Saint-Martin']
```

```python
user_latine = user[user['last_country_nm'].isin(countries_Latine)]
```

```python
user_latine.shape
```

```python
user_latine.last_locale.value_counts()
```

```python
#es_ES User only

user_latine=user_latine.loc[user_latine['last_locale'] == 'es_ES']
```

```python
user_latine['last_country_nm'].unique()
```

```python
user_latine['last_country_nm'].nunique()
```

```python
user_latine['last_country_nm'].value_counts()
```

```python
#Latine country data proportion 

test=(user_latine['last_country_nm'].value_counts()/user_latine['last_country_nm'].value_counts().sum())*100
round(test,1)
```

```python
user_latine.to_csv('/content/drive/MyDrive/01.project/mx/user_latine.csv.gz',index=False)
```

Divide into Mexico/Non Mexico Latine contries

```python
idx = user_latine.loc[user_latine['last_country_nm'] == 'Mexico'].index
user_latine_non_mx = user_latine.drop(idx, axis = 0)
```

```python
user_latine_non_mx['last_country_nm'].value_counts()
```

```python
#Latine country data proportion 

test=(user_latine_non_mx['last_country_nm'].value_counts()/user_latine_non_mx['last_country_nm'].value_counts().sum())*100
round(test,1)
```

```python
user_latine_non_mx.to_csv('/content/drive/MyDrive/01.project/mx/user_latine_non_mx.csv.gz',index=False)
```
