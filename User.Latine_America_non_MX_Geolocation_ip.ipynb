```python
#Latine top 5 country 

user_latine_Colombia=user_latine_non_mx.loc[user_latine_non_mx['last_country_nm'] == 'Colombia']
user_latine_Peru=user_latine_non_mx.loc[user_latine_non_mx['last_country_nm'] == 'Peru']
user_latine_Ecuador=user_latine_non_mx.loc[user_latine_non_mx['last_country_nm'] == 'Ecuador']
user_latine_Argentina=user_latine_non_mx.loc[user_latine_non_mx['last_country_nm'] == 'Argentina']
user_latine_Bolivia=user_latine_non_mx.loc[user_latine_non_mx['last_country_nm'] == 'Bolivia']
```

```python
#IP address imformation Extract

import os
import geoip2.database
import socket
import time

reader = geoip2.database.Reader('/content/drive/Shareddrives/Qanda/GeoLite2/GeoLite2-City.mmdb')
```

```python
#latitude_longtitude from IP Address_
for x in user_latine_Colombia.index:
    try:
        ip_add = str(user_latine_Colombia['last_ip_address'][x])
        response = reader.city(ip_add)
        user_latine_Colombia['country_name'][x] = response.country.name
        #user_latine_Colombia['city_name'][x] = response.city.name
        #user_latine_Colombia['latitude'][x] = response.location.latitude
        #user_latine_Colombia['longitude'][x] = response.location.longitude

    except Exception as msg:
        user_latine_Colombia['country_name'][x] = ""
        #user_latine_Colombia['city_name'][x] = ""
        #user_latine_Colombia['latitude'][x] = ""
        #user_latine_Colombia['longitude'][x] = ""
```
