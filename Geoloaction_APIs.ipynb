```
test_b=r'/content/drive/MyDrive/01.project/user_Brazil_mexico_ip.csv.gz'
test=pd.read_csv(test_b)
```

```
test.head(10)
```

```
import json
import urllib.request

GEO_IP_API_URL = 'http://ip-api.com/json/'
```

```
#city name from IP Address
for x in test.index:
    try:
        IP_TO_SEARCH = str(test['last_ip_address'][x])
        req = urllib.request.Request(GEO_IP_API_URL+IP_TO_SEARCH)
        response = urllib.request.urlopen(req).read()
        json_response = json.loads(response.decode('utf-8'))
        test['City_name'][x] = json_response['city']
        #print(x, "Finished,value is:",response.country.name)   
    except Exception as msg:
        test['City_name'][x] = ""
        #print(x," Finished with Error Msg:",msg)
```
